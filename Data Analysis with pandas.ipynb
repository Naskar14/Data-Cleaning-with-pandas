# Complete data cleaning using pandas

```python
df = pd.read_excel("C:/Users/Banashree/Downloads/Customer Call List.xlsx")
df
```

```python
#dropping the duplicate rows
df = df.drop_duplicates()
```

```python
df
```

```python
#dropping the useless columns
df = df.drop(columns = "Not_Useful_Column")
df
```

```python
df.shape
```

```python
#now need to clean up the column 'Last name'
df['Last_Name'] = df['Last_Name'].str.lstrip("...")
df['Last_Name'] = df['Last_Name'].str.lstrip("/")
df['Last_Name'] = df['Last_Name'].str.rstrip("_")
df['Last_Name']
```

```python
df['Last_Name'].str.strip("123._/")
```

```python
df['Last_Name'] = df['Last_Name'].str.strip("123._/")
df
```

```python
#cleaning up the phone number columns, the hardest part here. 1st I'll check the data type of it
df.info()
```

```python
#cleaning up all the special characters from this column
#df["Phone_Number"].str.replace('[^a-zA-Z0-9]','')
df["Phone_Number"] = df["Phone_Number"].str.replace('-','')
df["Phone_Number"] = df["Phone_Number"].str.replace('/','')
df["Phone_Number"] = df["Phone_Number"].str.replace('|','')
df["Phone_Number"]
```

```python
#putting into currect format
df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
```

```python
#getting the error because this function is not allowing me to do so when some of values are not a string. 
```

```python
df["Phone_Number"] = df["Phone_Number"].astype(str)
```

```python
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
```

```python
df
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('nan--', '')
df['Phone_Number'] = df['Phone_Number'].str.replace('Na--', '')
df
```

```python
#now trying to split the address column in different columns
df['Address'].str.split(',', expand=True)
```

```python
df[['street Address','State', 'Zip']]  = df['Address'].str.split(',', expand=True)
df
```

```python
#now changing the Paying Customer column either to yes or no

df['Paying Customer'] = df['Paying Customer'].str.replace('Yes', 'Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No', 'N')
df['Paying Customer']
```

```python
df = df.replace('N/a', '')
df = df.replace('NaN', '')
df
```

```python
df = df.fillna('')
df
```
